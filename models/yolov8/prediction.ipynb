```python
from ultralytics import YOLO

# Load a model
model = YOLO(r'C:\Users\SIDHARTH\Downloads\yolov8_original\content\runs\detect\train\weights\best.pt')  # pretrained YOLOv8n model
```

```python
import pandas as pd

prediction  = pd.DataFrame()


df = pd.read_csv("Test.csv")


for i in set(df.Path.values):
    result = model.predict(i)
    df1 = pd.DataFrame(result[0].boxes.xyxy)
    df1['Predicted_class'] = result[0].boxes.cls
    df1['Path'] = i
    prediction = pd.concat([prediction,df1])
    
    
```

```python
prediction = prediction.sort_values("Path")
```

```python
p = prediction.groupby("Path").apply(lambda x: x.tail(1)).reset_index(drop=True)
```

```python
p.head()
```

```python
Results = p.merge(df,on="Path")
```

```python
Results.head()
```

```python
from sklearn.metrics import classification_report,accuracy_score
accuracy_score(Results['ClassId'],Results['Predicted_class'])
```

```python
from sklearn.metrics import f1_score, mean_squared_error, precision_score, recall_score,accuracy_score,confusion_matrix

test_precision = precision_score(Results['ClassId'],Results['Predicted_class'], average='weighted')
test_recall = recall_score(Results['ClassId'],Results['Predicted_class'], average='weighted')
test_f1_score = f1_score(Results['ClassId'],Results['Predicted_class'], average = 'weighted')
```

```python
print("Test Precision : {0:.2f}".format(test_precision))
print("Test Recall : {0:.2f}".format(test_recall))
print("Test f1-score : {0:.2f}".format(test_f1_score))
```
